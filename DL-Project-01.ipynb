Appending all .csv files

```python
#Appending all csv files in us_census
us_census=pd.DataFrame()
for files in filesname:
    df=pd.read_csv(files)
    us_census=us_census.append(df,ignore_index=True)
```

```python
us_census.drop_duplicates('State',inplace =True)
us_census
```

```python
#Reseting Indexes
us_census['Unnamed:0']=us_census.index=(range(0,51))
```

```python
us_census
```

```python
#Removing Unnamed:0 column
us_census=us_census.drop('Unnamed:0',axis=1)
```

```python
us_census.head()
```

```python
#Splitting GenderPop
us_census[['Males','Females']]=us_census.GenderPop.str.split('_',expand=True)
```

```python
us_census=us_census.drop('GenderPop',axis=1)
```

```python
us_census.head()
```

```python
#Removing M and F from males and females column 
us_census['Males'] = us_census['Males'].str.replace('M', '').astype(int)
us_census['Females'] = us_census['Females'].str.replace('F', '0').astype(int)
     
```

```python
us_census['Females']=us_census['Females'].fillna(0)
```

```python
us_census.head()
```

```python
us_census['Hispanic'] = us_census['Hispanic'].str.replace('%','').astype(float)
us_census['White'] = us_census['White'].str.replace('%','').astype(float)
us_census['Black'] = us_census['Black'].str.replace('%','').astype(float)
us_census['Native'] = us_census['Native'].str.replace('%','').astype(float)
us_census['Asian'] = us_census['Asian'].str.replace('%','').astype(float)
us_census['Pacific'] = us_census['Pacific'].str.replace('%','').astype(float)
us_census['Income'] = us_census['Income'].str.replace('$','').astype(float)
```

```python
us_census.head()
```

```python
#After Updating checking datatypes
us_census.dtypes
```

```python
#Filling NAN values
us_census=us_census.fillna(0)
```

```python
us_census=us_census.round(decimals=6)
```

```python
us_census.head()
```
