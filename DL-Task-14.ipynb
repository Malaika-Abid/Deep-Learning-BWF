```python
import sqlite3
query = """
  CREATE TABLE test1
(a VARCHAR(20), b VARCHAR(20),
 c REAL, d INTEGER
  );"""
```

```python
con = sqlite3.connect('mydata.sqlite')
con.execute(query)
```

```python
con.commit()
```

```python
data = [('Atlanta', 'Georgia', 1.25, 6),
  ('Tallahassee', 'Florida', 2.6, 3),
  ('Sacramento', 'California', 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"
con.executemany(stmt, data)
```

```python
con.commit()
```

```python
cursor = con.execute('select * from test')
rows = cursor.fetchall()
```

```python
rows
```

```python
(cursor.description
```

```python
import pandas as pd 
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```

```python
#Example-02
import sqlite3
query=("""CREATE TABLE my_table4
(  id INTEGER ,
  name TEXT,
  age INTEGER)""");



```

```python
con = sqlite3.connect('mydata.sqlite')
con.execute(query)
```

```python
con.commit()
```

```python
data = [(1,'Alice', 25),
        (2,'Bob', 30),
        (3,'Charlie', 35)]

stmt = "INSERT INTO my_table4 VALUES (?, ? ,?)"
con.executemany(stmt, data)
```

```python
con.commit()
```

```python
cursor = con.execute('select * from my_table4')
rows = cursor.fetchall()
```

```python
rows
```

```python
(cursor.description)
```

```python
import pandas as pd 
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```
